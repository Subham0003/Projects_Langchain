```python
###lets read webbase documents
## Web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader=WebBaseLoader(web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
                     bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                         class_=("post-title","post-content","post-header")
                     ))
                     )
```

```python
loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=45)
final_document=text_splitter.split_documents(pdf_documents)
final_document
```

```python
speech=""
with open("speech.txt") as f:
    speech=f.read()


text_splitter=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=20)
text=text_splitter.create_documents([speech])
print(text[0])
print(text[1])
```

```python
##splitting through characteristic textsplitters

from langchain_text_splitters import CharacterTextSplitter
text_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=500,chunk_overlap=20)
final_document=text_splitter.split_documents(text_documents)
final_document
```

```python
import json
import requests
```

```python

json_data=requests.get("https://api.smith.langchain.com/openapi.json").json()
```

```python
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter
json_splitter=RecursiveJsonSplitter(max_chunk_size=100)

final_document=json_splitter.split_json(json_data)

print(final_document)
```

```python
for chunk in final_document[:3]:
    print(chunk)
```

```python
docs=json_splitter.create_documents(texts=[json_data])
for doc in docs[:3]:
    print(doc)
```
